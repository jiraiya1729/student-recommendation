```python
import pandas as pd
import json
import random
import string
import numpy as np
```

```python
def generate_unique_id(length=7):
    characters = string.ascii_lowercase + ''.join(map(str, range(1, 10)))
    return ''.join(random.choices(characters, k=length))
unique_ids = set()
for _ in range(len(df)):
  while True:
    new_id = generate_unique_id()
    if new_id not in unique_ids:
      unique_ids.add(new_id)
      break
df['unique_id'] = list(unique_ids)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(df['Story'])
```

```python
cosine_sim = np.dot(X, X.T).toarray()
```

```python
cosine_sim
```

```python
similarity_matrix_df = pd.DataFrame(cosine_sim)
similarity_matrix_df.to_csv('similarity.csv', index=False)
```

```python
top_11_indices = []

# Iterate over each row index in the cosine similarity matrix
for idx in range(cosine_sim.shape[0]):
    # Get the current row
    row = cosine_sim[idx, :]
    # Sort the indices based on the cosine similarity scores in descending order
    sorted_indices = np.argsort(row)[::-1]
    # Exclude the current student's own index (cosine similarity with itself will always be 1)
    sorted_indices = sorted_indices[sorted_indices != idx]
    # Take the top 11 indices
    top_11 = sorted_indices[:11]
    # Store the top 11 indices in the list
    top_11_indices.append(top_11)
```

```python
df_top_11_indices = pd.DataFrame(top_11_indices)
```
